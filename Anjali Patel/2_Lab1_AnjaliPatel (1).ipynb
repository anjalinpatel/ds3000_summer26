## Part 3.2: Manipulating Data (20 points)

1. Create a subset of the data set which (a) **includes** only train stations in Belgium and (b) **excludes** all train stations which are **not** in a city. Make sure to save this subset as a new data frame and print the first few rows of the data.
1. Use the `.describe()` function to produce summary statistics for the subset from the previous part. Create a markdown cell and explain:
    - What Series did the `.describe()` function run on? What Series did it not run on? What is the difference, and what does this mean the `.describe()` function is used for?

```python
df_belgium_train_stations = df_train_stations[
    (df_train_stations['country'] == 'BE') & 
    (df_train_stations['is_city'] == True)
    ]
print(df_belgium_train_stations.describe())
```

The .describe() function ran on the columns with numbers in them, which were latitude, longitude, and parent_station_id. It gave  stuff like the average, the smallest and biggest values, and the mean. It didn't run on the columns that had words in them or the ones with Boolean values. So .describe() is used to  see the basic math info about the number columns in your data, which helps you understand what the data looks like without having to look at every single row.
